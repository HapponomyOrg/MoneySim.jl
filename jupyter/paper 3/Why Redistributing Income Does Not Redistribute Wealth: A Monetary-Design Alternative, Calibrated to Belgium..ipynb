# Why Redistributing Income Does Not Redistribute Wealth: A Monetary-Design Alternative, Calibrated to Belgium.

# Abstract  
Inequality has grown across most of the world over recent decades (UN, 2020; OECD, 2015; OXFAM, 2025), and progress against poverty has stalled (World Bank Group, 2023) despite efforts from governments. Conventional redistributive instruments — progressive taxation, social transfers, welfare programs — operate as parameter adjustments within an existing monetary and fiscal architecture, and require continuous political maintenance against the upward drift of inequality they aim to correct. This paper argues that intervention at the level of the monetary system itself has higher structural leverage, because the dynamics that drive inequality — bargaining asymmetry, wealth-acquisition feedback, shock vulnerability, the conversion of wealth into political influence — operate through a common monetary medium, and it is on the distribution of that medium, monetary balances, that the present analysis and simulations operate.

The paper examines an Alternative Monetary System (AMS), which combines an unconditional guaranteed income with a demurrage charge on account balances, producing a continuous downward pressure on monetary accumulation, and a money stock that is bounded by design. AMS is calibrated to Belgian fiscal per capita data and simulated starting from the empirical Belgian wealth and income distributions. The simulations introduce a demurrage tax (demtax) layered on the stabilizing demurrage that is calibrated to fund the residual government expenditure remaining after the elimination of public-debt interest costs.

Because demtax operates on the bounded money stock that AMS produces, the demtax rate required to cover any specified expenditure target can be calculated directly from the architecture's parameters rather than searched for empirically. The simulations confirm that the calculated rate produces the targeted revenue while maintaining a stable money stock and substantially reducing monetary inequality. A counterfactual scenario funds the same expenditure through a conventional flow-coupled instrument — a progressive income tax — in place of the demtax. At the calibrated level of activity it covers the obligation, but its yield is tied to the volume of economic activity: as output contracts, coverage falls away, and a sustained decline drives the income tax to its rate ceiling while leaving the bulk of the obligation unmet, whereas the demtax raises the same revenue throughout the same decline.

Where the income tax stabilizes the income distribution but leaves monetary wealth concentrated, the demurrage compresses wealth regardless of the fiscal instrument layered on top — reaching comparable income equality without recourse to an income tax, and delivering the reduction in wealth concentration that flow-coupled taxation structurally cannot.

The behavioural responses that real agents would exhibit in such an architecture remain to be characterized through empirical methodologies beyond the scope of the present paper. Early evidence from a gaming-experiment study of an AMS-like environment is consistent with the direction theoretical argument predicts (Haantjes, 2024), but characterizing the magnitude and robustness of these effects would require larger-scale empirical work and longitudinal studies of actual implementations.

## 1. Introduction
Economic equality refers to the fair distribution of wealth, resources, and opportunities within a society, and is widely treated as a foundational element of social justice (Rawls, 2005; Valentini, 2012). The pursuit of economic equality is grounded in the principle that all individuals should have access to necessities, opportunities for advancement, and a reasonable standard of living, regardless of their socioeconomic background.

From a moral and ethical standpoint, the concept is closely linked to distributive justice and the fair allocation of resources and benefits within society. The desire for fairness and reduced inequality is not, however, only a philosophical commitment. Survey evidence indicates that populations consistently prefer more equal wealth distributions than the ones they actually live with (Norton and Ariely, 2011), and that a preference for fair distribution holds across cultural contexts even when the form of fairness varies (Starmans et al., 2017). Neurologic research finds that judgments of fair inequality recruit distinct neural networks, suggesting that fairness preferences are reflected at the level of neural processing (Cappelen et al., 2014).

Despite this widespread preference for fairness, inequality has grown across most of the world over recent decades. Reports from international institutions document the trend across regions and measurement approaches (OECD, 2015; Zucman, 2019; UN, 2020; OXFAM, 2025), and progress in reducing global poverty has stalled (World Bank Group, 2023). The disparity between income and wealth inequality is particularly pronounced: wealth inequality is consistently higher than income inequality (OECD, 2024), and the dynamics that produce wealth concentration appear to operate independently of, and more persistently than, the dynamics that produce income concentration. Recent work using Distributional National Accounts methodology suggests that inequality in Europe — long considered a region of relatively contained inequality — may be substantially higher than conventional measures indicate (Decoster and Hassan, 2023; Cantante, 2020). The conventional redistributive instruments deployed by states and international institutions over the post-war period — progressive taxation, social transfers, welfare programs, public provision of education and healthcare — have not arrested this trend, and on most measures inequality has continued to rise even where these instruments are most strongly implemented.

This persistence is not accidental. As Piketty (2022) argues, advancing and maintaining a decrease in inequality requires constant vigilance: when left unattended, inequality has a structural tendency to increase. From a systems-thinking perspective (Meadows, 2009), this implies that the socio-economic system contains one or more reinforcing feedback loops driving inequality upward, and that conventional redistributive instruments are operating against these loops rather than at their source. Left unchecked, reinforcing feedback loops are detrimental to any system, and it is therefore paramount that they be identified and that interventions be designed to counter their effects at the level of the dynamics that produce them, rather than at the level of the outcomes those dynamics generate.

This paper takes that systems-thinking framing seriously and asks where intervention should occur. It identifies the most prominent reinforcing feedback loops driving inequality in conventional monetary economies, drawing on evidence from economics, econophysics, psychology, and political science. It argues that intervention at the level of the monetary system has higher structural leverage than intervention at any of the alternative points commonly proposed in the heterodox literature, because the dynamics that drive inequality consistently operate through a common monetary medium. It then examines an Alternative Monetary System – henceforth AMS – a monetary architecture combining an unconditional guaranteed income with a demurrage charge on account balances, producing a money stock bounded by design and a continuous downward pressure on monetary accumulation.

The paper's central contribution is an examination of system dynamics: whether the proposed structural intervention changes the dynamics of wealth distribution and fiscal coverage in ways that are quantitatively different from what conventional instruments can achieve. To make this examination concrete, the paper presents a simulation calibrated with Belgian fiscal and demographic data as an illustrative scenario. The simulations introduce a demurrage tax (demtax) layered on the stabilizing demurrage, calibrated to fund government expenditure excluding interest on debt. Because demtax operates on the bounded money stock that AMS produces, the demtax rate required to cover any specified expenditure target can be calculated directly from the architecture's parameters rather than searched for empirically, a structural property that distinguishes demtax from conventional flow-coupled taxation. Counterfactual scenarios replace the demtax with conventional flow-coupled taxes. They are run both within the AMS architecture and within a conventional one, and test whether flow-coupled taxation can match what the demtax delivers.

The simulations demonstrate that the calculated demtax rate produces the targeted revenue exactly, while maintaining a bounded money stock. Within AMS the wealth outcome is driven by the medium and is robust to which fiscal instrument is layered on — though the stock-acting demtax compresses wealth inequality somewhat further than the income tax. The simulations further show that income tax alone does not reach wealth, even though it is an effective instrument to reduce income inequality, and the instrument fails to cover expenses in a scenario where GDP is declining. In that same scenario demtax holds its ground.

How real agents would respond to such an architecture, and what further effects those responses might have on the distribution, are discussed in Section 7 and given preliminary support by a small gaming study (Haantjes, 2024), but cannot be tested by the simulation methodology used here. The paper's findings therefore concern the mechanical dynamics of the proposed intervention; behavioural validation is flagged as a direction for future empirical work.

The remainder of the paper is structured as follows. Section 2 describes the methodology. Section 3 identifies the reinforcing feedback loops driving inequality in conventional economies. Section 4 develops the argument that intervention at the monetary level has the highest structural leverage among the available intervention points. Section 5 introduces AMS as a candidate mechanism. Section 6 presents the Belgian simulation and its results. Sections 7 and 8 discuss the implications and limitations of the findings, Section 9 sets out directions for further research, and Section 10 concludes.

## 2. Methodology
The paper combines a literature-based analysis with a computational simulation. The literature analysis is conducted in two stages. The first stage identifies the most prominent reinforcing feedback loops driving inequality in conventional monetary economies, drawing on findings from economics, econophysics, psychology, sociology, and political science. The second stage analyzes these loops to determine where structural intervention would have the highest leverage in the sense of Meadows (2009): the level at which a structural change to the system rules produces the largest and most durable effect on the dynamics those rules generate. This second stage is the methodological basis for the paper's central claim that intervention at the level of the monetary system is preferable to intervention at any of the alternative points proposed in the heterodox literature.
The computational simulations test whether AMS is feasible at present-day expenditure levels in a real economy. The simulations are implemented as an agent-based model using a yard-sale transaction mechanic, calibrated to Belgian fiscal data – total money stock, GDP, and total government expenditure. GDP is used as a time reference for the simulation, with one simulated year corresponding to a volume of yard-sale transactions equal to annual Belgian GDP, scaled to the number of agents.  

The simulations are run with the empirical Belgian wealth distribution scaled to M3, which tests whether AMS reduces existing inequality from a realistic starting point. Income cannot be drawn from the same data, because the joint distribution of wealth and income across individuals is not observed — the data fix each marginal distribution but not which income accrues to which wealth holder. For the first cycle income is therefore assigned in alignment with the initial wealth ranking, each agent receiving the income matching its wealth position; from the second cycle onward it is determined endogenously by the yard-sale dynamics. Because income becomes endogenous immediately and the runs span up to a century, this assignment affects only the first cycle and washes out: the steady-state relationship between wealth and income is a product of the dynamics, not of the starting assumption.  

The baseline is a pure yard-sale model with no AMS and no fiscal layer, which serves as a worst-case reference showing the unmitigated dynamics. The three intervention scenarios all share AMS as a common intervention — guaranteed income and stabilizing demurrage are active in each — and differ only in the fiscal mechanism applied on top: the pure-AMS scenario applies no additional taxation, isolating the redistributive effect of guaranteed income and demurrage alone; the demtax scenario applies a demurrage tax layered on the stabilizing demurrage, collected and immediately redistributed monthly; and the income tax scenario applies a progressive income tax calibrated to produce total government income equivalent to real Belgian government income, collected and redistributed monthly. Because all three intervention scenarios share AMS as a common intervention, the comparison between them isolates the effect of the fiscal mechanism alone, with the monetary architecture held constant.  

The simulations track seven outcomes. Two describe the distribution of monetary wealth: the shares held by percentile bands — the top 0.1%, top 10%, upper-middle 40%, lower-middle 40%, bottom 10%, and bottom 1% — and the wealth Gini coefficient. Two describe the income distribution in the same terms, the income percentile shares and the income Gini, where income is measured in two ways: gross monetary inflow — the sum of incoming transactions, guaranteed income, and any redistributed government revenue received by an agent over a simulated year, corresponding to the income concept used in conventional inequality statistics — and transactional income — the sum of incoming agent-to-agent transactions alone, excluding guaranteed income and redistribution — which isolates the inequality generated by economic activity independently of the AMS interventions. The fifth outcome is the ratio of the wealth Gini to the income Gini, which captures the relationship between the two distributions and, in particular, the extent to which wealth inequality exceeds income inequality. The sixth is the total money stock, as a measure of monetary stability. The seventh is the stock of public debt, which begins at an initial level and accumulates through deficit spending; it serves as the measure of fiscal coverage, remaining flat when the fiscal layer meets expenditure in full and building up when expenditure is met by deficit instead.

## 3. Identifying reinforcing feedback loops
Much of the inequality literature focuses on income and consumption inequality, with the more difficult question of wealth inequality treated less systematically. The two are connected. When part of the population has income that exceeds the cost of living, the surplus is available to acquire capital, and the capital generates further income, which can be reinvested in further capital. Income inequality therefore tends to translate into wealth inequality over time, and wealth inequality tends to reinforce income inequality through the returns it generates. This reinforcing feedback loop is known as the Matthew effect, and it is the most general formulation of the dynamic this section is concerned with: inequality, once present, tends to amplify itself through mechanisms embedded in the way economies allocate resources.  

Piketty (2022) has argued that the underlying driver of this amplification in modern capitalist economies is the gap between the rate of return on capital ($r$) and the rate of economic growth ($g$). When r>gr > gr>g, wealth held as capital grows faster than the economy as a whole, and wealth concentrates over time at the top of the distribution. Piketty's empirical work suggests that r>g r > g r>g has been the historical norm rather than the exception, with the post-war decades of compressed inequality in developed economies being a temporary deviation produced by specific political and fiscal conditions that no longer hold. Subsequent work has tested and refined this thesis (Madsen, 2017), and the central observation that capital accumulation tends to outpace economic growth in most periods has been broadly supported even where its policy implications remain contested.  

The yard-sale model, introduced in econophysics by Chakraborti (2002) and named in a review by Hayes (2002), provides a formal treatment of the wealth-concentration dynamic. In the model, random pairs of agents engage in transactions in which the amount exchanged is a fixed proportion of the poorer agent's wealth and the direction of transfer is determined by a fair coin flip. Despite the apparent symmetry of these rules, the model invariably drives wealth toward complete concentration in the hands of a single agent in the time-asymptotic limit. Boghosian (2014a, 2014b) derived the kinetic equations governing this process, and Boghosian, Johnson, and Marcq (2015) proved that inequality in this model, measured by the Gini coefficient, can only increase over time, so that the system is driven with mathematical certainty toward complete concentration rather than merely tending toward it. The mechanism producing this result is subtle. Although the coin flip is fair and each agent is equally likely to win any given transaction, the wealthier agent risks a smaller fraction of total wealth in each exchange, because the stake is set by the poorer agent's holdings. It is the asymmetry of stakes, not any asymmetry of probability, that produces the concentration.  

Real-world wealth distributions are more compressed than the bare yard-sale model predicts, indicating that the pure concentration dynamic is offset by other mechanisms in actual economies. Extensions of the model that incorporate redistribution, a wealth-attained advantage favouring richer agents in exchange, and the possibility of negative wealth have been shown to reproduce empirical wealth distributions for several economies to within a fraction of a percent (Boghosian et al., 2017; Li et al., 2019; Boghosian et al., 2022). These extensions establish that the dynamics captured by the model are not merely theoretical artefacts but operate observably in real economies. More recent work has examined the model on networks rather than fully connected populations, finding that the global wealth condensation of the original model is replaced by a form of local condensation, which suggests that the fully connected formulation represents an upper bound on concentration relative to more realistic interaction structures (Vazquez Von Bibow and Perotti, 2025).  

The yard-sale model is formally a model of wealth exchange in the broad sense, with its state variable encompassing any economically transferable holding. The relationship between this wealth-level formalism and the monetary intervention proposed in this paper is addressed in Section 4.3, where the case is made for treating the formal wealth variable through its monetary expression for the purposes of the intervention examined here.  
The wealth-attained advantage that the extended model formalises has a straightforward real-world correlate in the bargaining asymmetry between richer and poorer parties. Wealthy actors enter economic transactions with lower proportional stakes: a given monetary loss represents a smaller fraction of their resources, and they can decline any particular exchange without serious consequence, while poorer actors often cannot. This asymmetry translates into negotiating leverage across the full range of economic interactions, from wage bargaining between employers and workers to price and contract negotiations between buyers and sellers and between landlords and tenants (Jeong, 2013). In each case the party with greater wealth holds an advantage that derives not from the economic substance of the exchange but from the differential cost of failing to reach agreement. Repeated across the transactions of an economy, this asymmetry produces a continuous transfer on terms set by relative wealth rather than by the underlying value exchanged: wealth confers bargaining power, bargaining power secures favourable terms, and favourable terms produce further wealth.  

Socio-economic shocks, such as the 2008 financial crisis and the COVID-19 pandemic, affect vulnerable groups asymmetrically (Pfeffer et al., 2013; Compton et al., 2018; Van Bavel and Scheffer, 2021; Perry et al., 2021; Popyk and Pustułka, 2023), widening the wealth and opportunity gap during and after each episode. The asymmetry has both a material and a cognitive component. Materially, affluent groups hold larger buffers that absorb income loss and asset depreciation without lasting damage to their position. Cognitively, the financial scarcity that shocks impose on those with smaller buffers itself degrades decision-making, narrowing attention and foreshortening time horizons in ways that compound the material disadvantage (Mullainathan and Shafir, 2021). Systemic analysis has identified 151 banking crises, 236 currency crises, and 75 sovereign debt crises worldwide between 1970 and 2017 (Laeven and Valencia, 2018), establishing that such shocks are recurrent features of the financial architecture rather than exceptional events. Each episode redistributes wealth from those with smaller buffers to those with larger ones, and the cumulative effect across repeated shocks over a generation is substantial.  

The recurrence of these crises is not accidental. The conventional monetary architecture creates money through credit issuance, with commercial banks creating deposits when they extend loans, and this mechanism channels newly created money toward those who can already offer collateral, which is to say toward existing wealth (Cincotti et al., 2010; Kuypers et al., 2021). Historical analysis indicates that interest-bearing debt in slow-growing or non-growing economies tends to produce unpayable debt accumulations that resolve only through default, forgiveness, or political intervention (Hartley and Kallis, 2021). The credit-money architecture is therefore not merely a setting within which inequality arises from other dynamics; it is itself a dynamic that produces inequality through the structural relationship between collateral, credit access, and money creation. The monetary policies implemented by central banks operate within this architecture and inherit its bias. Contractionary policy appears to worsen income inequality (Coibion et al., 2017; Mumtaz and Theophilopoulou, 2017; Creel and El Herradi, 2024), and quantitative easing also appears to increase it, most plausibly because money created through asset purchases accrues first to those who already hold financial assets (Mumtaz and Theophilopoulou, 2017). The instability and inequality associated with the credit-money architecture are, on this reading, consequences of the mechanism by which money is created and allocated rather than failures of particular policies.  

Wealth concentration also produces political concentration, which in turn produces policy that further entrenches wealth concentration. Empirical work on the relationship between wealth and political influence in democratic systems has consistently found that the policy preferences of wealthy citizens correlate with policy outcomes while the preferences of less wealthy citizens have little independent effect (Gilens and Page, 2014; Page et al., 2013). The personal economic interests of officeholders correlate with the interventions they pursue (Tahoun and Van Lent, 2019), and individual wealth has been shown to translate into political power through channels outside formal representation (Rossi, 2014). The mechanisms are familiar: campaign finance, lobbying, ownership of media, social ties between economic and political elites, and movement between regulatory bodies and the industries they oversee. Together they form a loop in which wealth purchases influence, influence shapes policy, policy shapes the distribution of returns, and the distribution of returns reinforces the wealth that began the cycle. This loop is particularly resistant to democratic correction, because the political power required to correct it is the same power the loop concentrates in those who benefit from it.
The persistence of concentration is supported, finally, by ideological dynamics that shape how inequality is perceived and justified. More unequal societies tend to exhibit stronger individual belief in meritocracy, the conviction that economic outcomes reflect individual merit rather than structural conditions (Mijs, 2019; Goya-Tocchetto and Payne, 2022), and economic inequality has been shown to intensify the desire for wealth and status, producing dispositions that feed back into the dynamics generating inequality (Wang et al., 2023). Sandel (2020) traces the political ascent of meritocratic ideology through the late twentieth century and argues that it has furnished moral cover for policies that increased inequality over the same period. The relationship between meritocratic belief and material inequality is bidirectional: belief in meritocracy legitimates the policies and behaviours that produce inequality, while the resulting inequality makes meritocratic explanations of individual outcomes more cognitively available. The ideological loop is not separate from the material loops examined above but operates alongside them, supplying the legitimation that makes them politically sustainable.  

The feedback loops identified in this section share several structural features. Each operates through behavioural and institutional responses to material conditions rather than through any single deliberate mechanism. Each is self-reinforcing, in that the outcomes it produces feed back into the conditions that generated them. And each resists piecemeal intervention: addressing the bargaining asymmetry through specific labour-market reform leaves the underlying wealth distribution intact, addressing the political-influence loop through campaign-finance reform leaves intact the wealth that purchases influence, and addressing meritocratic ideology through education or public discourse leaves intact the material conditions that continually regenerate it. These observations motivate the question the next section takes up: given the structure of the loops, where would intervention have the highest leverage, and why?

## 4. Why intervene at the monetary level
Section 3 identified a set of reinforcing feedback loops that drive the persistence and growth of inequality in conventional monetary economies. Identifying the loops, however, does not by itself indicate where intervention should occur. The same diagnosis is compatible with a wide range of proposed remedies — labour market reform, progressive wealth taxation, expanded public provision, ownership reform, behavioural interventions — each of which has its own constituency in the heterodox literature, and each of which addresses some subset of the observed dynamics. This section argues that intervention at the level of the monetary system has higher structural leverage than intervention at any of these alternative points, and that the dynamics identified in Section 3 are most efficiently addressed there. The argument proceeds in three steps: first, that behaviour is downstream of context and therefore an inefficient target of intervention; second, that interventions on system structure have higher leverage than interventions on system parameters; and third, that money is the substrate through which the dynamics of Section 3 most consistently operate.

### 4.1 Behaviour is downstream of context
The feedback loops examined in the previous section are, at their proximate level, behavioural. Wealthy actors use their bargaining advantage; poorer actors accept worse terms; people in financial scarcity make decisions under cognitive load; populations in highly unequal societies adopt meritocratic beliefs that legitimate the inequality they experience. Each of these is a description of human behaviour, and it is tempting to conclude that the appropriate intervention is therefore to modify the behaviour — through education, moral exhortation, voluntary codes of conduct, or institutional norms.
This conclusion misreads the relationship between behaviour and the conditions that produce it. The distribution of human dispositions — cooperative and competitive, generous and acquisitive, trusting and suspicious — has not changed materially over the historical timescales in which dramatically different distributional outcomes have been observed. What has changed is the context within which those dispositions are expressed and selected for. The same human capacities that produce extreme inequality under one set of institutional and material conditions produce relative equality under another. The variable that explains the difference in outcomes is therefore not the human material but the context in which it operates. This view has analogues in the econophysics literature, where Boghosian and colleagues (2022) draw a parallel between the dispersion of wealth among economic agents and the dispersion of kinetic energy among molecules in a gas: in both cases, the dispersion arises from the dynamics of interaction rather than from intrinsic differences between the interacting entities.

The literature cited in Section 3 supports this reading directly. Mullainathan and Shafir (2021) show that financial scarcity itself produces the cognitive patterns — tunnel vision, foreshortened time horizons, reduced strategic capacity — that are commonly attributed to the dispositions of those experiencing poverty. The same individuals, placed in conditions of financial sufficiency, exhibit different cognitive patterns. The bargaining-advantage literature (Jeong, 2013) describes a behavioural asymmetry that is generated by the underlying wealth distribution rather than by any difference in disposition between rich and poor actors.
The work on meritocratic belief in unequal societies (Mijs, 2019; Goya-Tocchetto and Payne, 2022; Wang et al., 2023) shows that ideological commitments themselves track contextual conditions. Meritocratic belief and material inequality stand in a mutually reinforcing relationship: belief in meritocracy legitimates the policies and behaviours that produce inequality, while the resulting inequality in turn makes meritocratic explanations for individual outcomes more cognitively salient. The post-1980 period in most OECD economies illustrates the loop in operation, with the rise of meritocratic ideology (Sandel, 2020) and the policy shifts it justified coinciding with, and reinforcing, a sustained increase in income and wealth inequality. What matters for the present argument is not the direction of causation but the fact of co-variation: ideological commitments to meritocracy are not invariant features of human disposition but emerge under specific structural conditions, which means they too are downstream of context rather than upstream of it.
Behavioural intervention, then, attempts to push against a context that continuously regenerates the behaviours it targets. The intervention may produce short-term effects but cannot durably alter the underlying dynamic, because the context that produced the behaviour remains in place to produce it again. Effective intervention must operate at the level of context.

Mainstream macroeconomics has begun to acknowledge that rational-agent assumptions produce unrealistic predictions about aggregate behavior, and recent work in behavioral New Keynesian modeling has incorporated bounded rationality and cognitive discounting as stable aggregate features of populations (Gabaix, 2020; Pfäuti and Seyrich, 2022). These developments represent a partial convergence with the position taken here — that certain behavioral tendencies are stable across populations and can be treated as parameters of economic models — but two important distinctions remain. The first is that the behavioral elements incorporated in this literature are narrower than the broader claim developed above: they capture specific cognitive limitations, such as myopia about distant events or heuristic decision-making, rather than the general observation that behavioral expression tracks contextual conditions. The second is that the behavioral frameworks developed in this literature are used to refine policy design within the existing monetary and fiscal architecture, producing modified Taylor rules, adjusted forward-guidance strategies, and recalibrated inflation targets. The architecture itself — the structural rules governing money creation, credit, and fiscal flows — remains outside the scope of these models. The behavioral New Keynesian literature therefore treats behavioral realism as a constraint on parameter-level policy interventions, rather than as a reason to rethink the architecture within which policy operates. The argument of this paper runs in the other direction: if behavior is downstream of context and context is structural, then structural intervention has higher leverage than parameter intervention, and the behavioral realities the BNK literature acknowledges are themselves reasons to examine the architecture rather than to refine policy within it.

### 4.2 Leverage and the choice of intervention point
Once intervention shifts from behaviour to context, a further question arises: which contextual level is the appropriate target? Donella Meadows's framework of leverage points (Meadows, 2009) offers a way to address this question. Meadows distinguishes between interventions on system parameters — tax rates, transfer levels, sectoral subsidies, regulatory thresholds — and interventions on system structure: the rules by which the system operates, the information flows that govern its behaviour, and the goals and paradigms within which it is organized. Parameter interventions are easier to enact and easier to reverse, but they operate within a system whose structure continues to generate the conditions the parameters are adjusted to address. Structural interventions are harder to enact and harder to reverse, but they change the conditions within which downstream dynamics play out, and therefore have larger and more durable effects per unit of intervention.
Conventional redistributive instruments — progressive taxation, social transfers, welfare programs, minimum wage legislation — are parameter interventions in this sense. They adjust the magnitudes of flows within an existing monetary and fiscal architecture without altering the architecture itself. This is why they require what Piketty (2022) describes as constant vigilance: the underlying dynamics that produce inequality continue to operate, and the parameter adjustments must be continuously maintained, defended, and recalibrated against the upward drift they are designed to counteract. When political conditions weaken the maintenance, inequality reasserts itself, as the post-1980 period in most OECD economies demonstrates.
A structural intervention, by contrast, would alter the rules by which the dynamics themselves operate. Rather than offsetting the upward drift of inequality through continuous redistributive effort, it would change the conditions under which that drift is generated in the first place. The advantage of such an intervention is not that it is easier to implement — structural change is generally harder than parameter change — but that, once implemented, it requires substantially less ongoing political effort to maintain, because the dynamics it addresses are no longer generating the conditions it is designed to correct. The question is therefore which structural level offers the highest leverage for the dynamics identified in Section 3.  

### 4.3 Money as the medium of distributional dynamics
The dynamics examined in Section 3 are heterogeneous in their immediate description — bargaining asymmetry, wealth-acquisition feedback, shock vulnerability, monetary policy effects, ideological reinforcement of inequality — but they share a common operating medium. Each, traced from its proximate behavioural expression to the underlying mechanism through which it produces distributional effects, runs through money. Money here is the medium in which value is denominated and through which it flows; the monetary system is the set of rules governing how that medium is created, destroyed, and transferred. The dynamics of Section 3 run through the medium, but they are shaped by the system, and it is the system that an intervention alters.  

The Matthew effect operates through the differential capacity of monetary surplus to be converted into income-generating assets: those with monetary income above subsistence can acquire capital, those without cannot, and the resulting capital generates further monetary income. The yard-sale dynamic, as a formal model, operates on monetary transfers between agents; its real-world correlate operates on the asymmetric stakes that monetary loss represents for actors at different points in the wealth distribution. Bargaining advantage between rich and poor actors is generated by the asymmetric monetary stakes of the transactions they enter: a given monetary loss represents a smaller proportional cost to the wealthy actor and a larger proportional cost to the poor one, which translates directly into negotiating leverage. Socio-economic shocks transmit through monetary channels: loss of employment income, depletion of monetary savings, restriction of credit access. The effects of monetary policy on inequality are by definition monetary. Even the ideological dynamics — the meritocratic legitimation of inequality, the conversion of wealth into political influence documented by Gilens and Page (2014) — operate through the monetary medium, since the wealth that generates meritocratic belief and political leverage is itself held and transferred in monetary form.  

This is not to claim that the monetary system is the only point at which distributional dynamics can be addressed; other systems govern particular dynamics and could be targeted in their own right. What distinguishes money is that it is the medium through which all of the dynamics identified in Section 3 traverse: an intervention on the system that governs money therefore reaches every one of those dynamics, whereas an intervention on any other system reaches only the dynamics that operate within its particular domain. Sector-specific or domain-specific interventions — labour market reform, housing policy, healthcare provision, ownership reform — address particular dynamics within particular domains, but each operates within the existing monetary architecture and inherits its dynamics. Wealth taxation, for instance, is an intervention on accumulated monetary wealth using the existing fiscal apparatus; it does not alter the dynamics that produced the accumulation in the first place, and is subject to the same continuous-maintenance requirement as other parameter interventions. The same is true of every other intervention that operates on monetary outcomes without altering the rules of the monetary system itself. There are few problems in the contemporary distributional landscape that can be analyzed without reference to money, and there are correspondingly few interventions on those problems whose effects are not mediated by the monetary architecture within which they are implemented.  

This argument has a direct bearing on the formal model used to study these dynamics. The yard-sale model, introduced in Section 3 and formally analyzed in a series of papers by Boghosian and colleagues (Boghosian, 2014a, 2014b; Boghosian et al., 2015, 2017, 2022), is a model of wealth exchange in the broad sense — its state variable encompasses any economically transferable holding, monetary or otherwise. The model's central result, that unmitigated yard-sale dynamics drive inequality toward complete concentration in the time-asymptotic limit (Boghosian et al., 2015), is therefore formally a result about wealth concentration, not specifically about monetary concentration. The formalism nevertheless treats wealth as a scalar quantity expressed in a single unit, which presupposes a monetary valuation of the underlying holdings: in the analysis of Boghosian et al. (2022), wealth is explicitly written as the sum of asset quantities multiplied by their prices. The state variable in the yard-sale model is therefore not the physical inventory of an agent's holdings but the monetarily valued sum of those holdings.
For the purpose of this paper, we take the further step of treating this monetarily valued scalar as the operative variable for a monetary intervention, on the grounds established above: wealth participates in distributional dynamics through its monetary expression, and the dynamics the yard-sale model describes are precisely the dynamics that operate on wealth in its monetarily active form. This is a scope-limiting interpretive choice rather than a claim about the formal model: the model's wealth variable retains its broader formal meaning, and this paper operationalizes it for the specific purpose of examining a monetary intervention on the dynamics it describes.  

The methodological corollary is that wealth that remains non-liquid is economically inert with respect to the dynamics under study. A share portfolio or a property holding does not, in itself, generate the feedback loops identified in Section 3. It is only when wealth is mobilized — through sale, liquidation, or use as collateral for credit — that it enters the circuits of economic activity that drive inequality. At that moment of mobilization, it takes the form of money. This holds even in cases where liquidation is avoided. When an asset is used as collateral to secure a loan — as occurs routinely among high-net-worth individuals — the economically active agent is not the asset itself but the liquidity it enables. The asset functions as a condition of possibility; the money is what flows, what purchases, what compounds. Even in informal economies, some form of currency mediates exchange; barter constitutes a negligible fraction of total transactions. Tracking monetary flows therefore captures the dynamics under study with sufficient fidelity to assess interventions that act on those flows. The limits of this assumption are examined in the limitations section.  

### 4.4 Implications for the choice of mechanism  
The argument of this section can be summarized as follows. The dynamics identified in Section 3 are behavioural at their proximate level but contextual in their generation; intervention on behaviour is therefore inefficient, and intervention should target context. Among the contextual levels at which intervention is possible, structural interventions on the rules of the system have higher leverage than parameter adjustments within the existing system, because the latter must be continuously maintained against dynamics that the former would alter at their source. Among the structural levels at which intervention is possible, the monetary system is distinguished by governing money, the medium through which the dynamics of Section 3 consistently operate, regardless of the domain in which their effects appear.  
If this argument holds, the question becomes what kind of monetary intervention would produce structurally redistributive effects without requiring continuous political maintenance. The next section introduces such a mechanism.

## 5 Proposition: a monetary mechanism for structural redistribution
Section 4 argued that intervention on the monetary system has higher structural leverage than intervention at the level of behaviour or of parameters within the existing fiscal apparatus, because money is the medium through which the dynamics identified in Section 3 consistently operate. That argument is silent, however, on what such an intervention would look like. This section introduces the alternative monetary architecture examined in this paper, henceforth AMS (Alternative Monetary System), which integrates redistribution into money creation and destruction itself, and which this paper proposes as a candidate structural intervention. The description here is compact; the convergence and boundedness properties used in what follows are set out in Section 5.1; a fuller formal treatment is developed separately.  

### 5.1 Core mechanism
AMS combines two long-standing concepts and operates them as the dual sides of a single money creation and destruction process. The first is an unconditional periodic payment to every individual, widely discussed in the contemporary literature under the heading of universal basic income (UBI; Van Parijs and Vanderborght, 2017). This paper uses the term guaranteed income (GI) rather than UBI, on the grounds that "basic" prejudges the level at which the income is set, whereas GI leaves the level open as a parameter. The second concept is demurrage: a holding fee charged at regular intervals on monetary balances, proposed by Silvio Gesell in the early twentieth century as a means of discouraging hoarding and accelerating monetary circulation (Gesell, 1958; Champ, 2008). Although demurrage has remained marginal in mainstream monetary theory, it received a brief endorsement from Keynes in the General Theory and has more recently attracted renewed attention in heterodox and ecological economics (Svensson and Westermark, 2020).  

In AMS, each living individual holds a single account eligible to receive a guaranteed income GI. At the end of every period demurrage is charged, at a percentage rate d on the balance of every account, eligible or not, and the collected demurrage is destroyed; the guaranteed income is then deposited into each eligible account, signalling the start of the next period. The GI is created ex nihilo by the issuing authority; it is not funded by taxation and is not anyone's liability. This order is not incidental. Charging demurrage before depositing the GI ensures that the income deposited at each boundary escapes that boundary's demurrage charge, which is what yields the clean convergence point derived below; were the GI deposited first and demurrage then applied to the combined balance, part of each period's income would be lost to demurrage immediately and the convergence point would fall to GI(1 − d)/d rather than GI/d.  

The combination produces a system in which each account converges toward a calculable equilibrium balance. For an account that receives only the guaranteed income, with no other net income or expenditure, this point is telo(GI, d) = GI/d, the balance at which periodic demurrage exactly offsets the periodic GI. More generally, for an account with a net periodic income I — the guaranteed income plus any income earned through transactions, less any expenditure — the convergence point is telo(I, d) = I/d. An account whose holder earns net income beyond the GI therefore settles at a higher equilibrium balance, and one who runs a net deficit settles at a lower one. The equilibrium distribution of balances across accounts thus tracks the distribution of net income rather than accumulating independently of it, a property examined empirically in Section 6.  

The total money stock nevertheless remains bounded. Because transactions are conservative — every unit of transactional income to one account is a unit of expenditure from another — net transactional income sums to zero across all accounts, and the total money stock converges to |A_GI| · telo(GI, d) = |A_GI| · GI/d, where |A_GI| is the number of eligible accounts. This bound holds regardless of the distribution of money across accounts and regardless of the volume of transactions between them: transfers redistribute the demurrage burden across accounts but do not change its total. The money stock is therefore bounded by design rather than by intervention.
Two properties of this mechanism are central to the argument that follows. First, the system produces continuous downward pressure on monetary accumulation: an account holding a balance above its convergence point loses more to demurrage than it receives in income, while an account below its convergence point gains, and the further an account sits above that point, the stronger the pressure. This is the inverse of the yard-sale dynamic identified in Section 3, in which wealth concentration is self-reinforcing; under AMS, concentration is self-attenuating. Second, the mechanism is parameterized by GI and d alone, both set at the level of monetary policy rather than fiscal negotiation. The redistributive effect therefore operates with the same insulation from short-term political dynamics as conventional monetary policy.  

### 5.2 From mechanism to fiscal architecture
The core mechanism described above is concerned only with money creation and destruction. It does not, in itself, fund government expenditure. To examine AMS in a real economy, the model must be extended with a fiscal layer. This paper introduces a demurrage tax (henceforth demtax): an additional rate, layered on top of the stabilizing demurrage, whose proceeds are not destroyed but flow to government as tax revenue. The demtax operates on the same base as demurrage — account balances at the start of each period — and is collected by the same mechanism. Where demurrage maintains the boundedness of the money stock, demtax funds the operations of the state.  

This fiscal layer raises the question of which functions the state must continue to perform under AMS. Several categories of current government expenditure exist as responses to features of the conventional monetary architecture rather than as services the state provides for their own sake.  

Unemployment support compensates for the precarity of income in a system where wage labour is the primary route to subsistence; under AMS, the guaranteed income absorbs this function directly, since it provides an income floor that does not depend on employment status. Pensions require more care. They too provide income in the absence of a baseline guarantee, but current pension entitlements typically exceed the guaranteed income, so the GI absorbs only the baseline of the pension function rather than the whole of it. The simulation accordingly retains a residual pension obligation: for the share of the population above retirement age, the state tops up the guaranteed income to the level of the average pension, at a per-period cost of N × p₆₅ × (pension_avg − GI), where p₆₅ is the share of the population aged 65 and over and pension_avg is the average pension. This residual is included in the expenditure the demtax must fund. The GI thus absorbs the portion of income-security provision that a baseline income can cover, while the excess of existing pensions over that baseline remains an explicit fiscal obligation rather than being assumed away. Consistent with the homogeneous-population simplification described in Section 2, this residual enters as an aggregate expenditure line rather than as a payment targeted to retirees; the simulation therefore captures the fiscal cost of the obligation while not modelling its distribution across age groups, which is a parameter-level matter that does not bear on the system dynamics under study.  

Interest on public debt warrants more careful treatment. Under AMS the state is not prevented from borrowing, but the conditions that produce positive interest rates on government debt dissolve. A holder whose balance sits above their telo pays both demurrage and demtax on the excess each period. Lending that excess to the state reduces their account balance and with it their liability to both charges, since the money lent is no longer in an account subject to either. Lending is therefore a gain relative to holding as long as the interest on the loan exceeds −(d + d_t), so zero and negative interest become rational on the lender's side, and the equilibrium interest rate on public debt collapses toward zero and below, with a floor at −(d + d_t). 

This leaves a benign form of deficit spending available to the state. It can issue bonds to account holders with balances above their telo, who have a standing incentive to take them up rather than absorb demurrage and demtax on idle balances; because the borrowed money moves from private accounts to the state and is spent back into private accounts, the operation is money-neutral and does not disturb the bounded money stock. Nor does it erode the demtax base: demtax is levied on the total of account balances, which the bounded money stock fixes, so lending merely shifts the money — and its demtax liability — from the lender's account to the accounts of those who receive the state's spending, leaving aggregate demtax revenue unchanged. Because the bonds can carry zero or negative interest, the borrowing accumulates no positive interest cost and cannot produce the compounding debt-service dynamics of conventional public debt. This is distinct from money-creating deficit spending, which AMS forecloses by reserving money creation to the guaranteed-income mechanism: the state cannot finance a deficit by creating money, but it can finance one by borrowing money that already exists, at a cost bounded below by the combined demurrage and demtax floor. AMS does not abolish public debt; it removes its cost, and the interest line in current government expenditure can be treated as structurally absent rather than as a burden the state must continue to bear. The simulations in this paper do not model bond financing — the fiscal layer is funded entirely by demtax — which is the more demanding assumption, since it requires demtax to cover the full residual expenditure without recourse to borrowing and credits AMS with none of the flexibility that negative-cost borrowing would provide.  

The fiscal layer examined in this paper relies on demtax alone, a levy on the stock of monetary balances. Conventional consumption taxation such as value-added tax could be retained alongside it without modification. Because VAT is levied on transactions, its revenue scales with the volume of economic activity, so combining it with demtax would give the state a revenue base that is partly stock-based and partly flow-based. The simulations omit VAT, both because its effect is straightforward to anticipate — it raises revenue in proportion to GDP — and because the question of interest here is whether a purely stock-based levy can fund the state on its own. The additional fiscal headroom a flow-based instrument would provide is noted but not pursued.

## 6 Inequality simulations
The runs in this section share one calibration — the Belgian money stock, the agent population, and the annual expenditure obligation fixed earlier — and differ only in controlled ways, so that each difference in outcome can be traced to the single feature that was changed. The obligation to be funded is held constant across every run that carries a fiscal layer, at full coverage, so that the systems are compared while funding the same expenditure rather than while applying the same tax rate; what varies, in turn, is the presence of the monetary substrate, the instrument used to raise the revenue, and the level of activity in the economy. The design is meant to keep apart two questions that are easily run together. Whether the wealth distribution compresses under the alternative system because of its monetary substrate — the bounded stock produced by the guaranteed income and demurrage — or because of the particular tax levied alongside it is answered by holding one fixed and varying the other. And whether the case for taxing the monetary stock is distributional or structural is settled not by the steady-state distributions, which establish only the weaker claim, but by what becomes of revenue as activity falls away. For each run we report the wealth and income distributions, as percentile shares and as Gini coefficients, the ratio between the two — in which the paper's central distinction between governing a flow and governing a stock becomes visible — and the trajectories of the money stock and the public debt.

The wealth Gini of 0.51 reflects the interaction of a steadily expanding money stock with a fixed volume of real activity: the equal, per-capita character of deficit-financed injection compresses the stock, while the accompanying fall in velocity raises measured concentration in the income flow, and together they pull the coefficient toward a value that is an artefact of the monetary expansion rather than a property of the fiscal instrument. The baseline run removes every corrective and lets the exchange model run alone. There is no monetary substrate and no fiscal layer: agents transact under the same multiplicative rule as in every other run, but nothing is set against the concentration that rule produces. Calibrated to the same starting distribution, with a wealth Gini of 0.61, the system does not hold there. Wealth pools steadily into fewer hands — the Gini passes 0.9 within five years and reaches essentially one, a single holder of the whole stock, by year ten. As it does, the economy stops: with almost nothing left in most hands to trade with, the volume of exchange drains away and output with it, falling to zero by year fourteen. The run is carried to year thirty, but nothing moves after the collapse: the concentrated state is absorbing.
This is the yard-sale dynamic, whose drift to complete oligarchy is established analytically in the literature [Boghosian et al.]; the run reproduces it on the Belgian calibration and adds two points that bear on what follows. The money stock is constant throughout — none is created or destroyed here — so the concentration is a pooling of a fixed stock, not an inflation. And the terminal state is not merely unequal but inoperative: an economy in which one agent holds everything has no transactions left to conduct, so the unconstrained dynamic does not come to rest at a working if lopsided distribution but at the point where exchange itself stops. Its only attractor is the one at which the economy ceases to function. Every distribution in the runs that follow is therefore the product of some force placed against this attractor, and the baseline fixes the reference at one end of the scale — complete concentration — against which those forces are read.

The first scenario strips out the fiscal layer entirely: agents receive the guaranteed income and pay demurrage, but no tax is levied and no public expenditure is funded. It is not offered as a policy — a monetary system carrying no public finance is not a complete one — but as the baseline against which every later run is read, isolating what the monetary substrate does on its own before any tax revenue is redistributed on top of it.
Two results follow. The money stock settles at a level fixed by the monetary parameters alone. Since the guaranteed income injects N·GI each period while demurrage withdraws a fraction d of all holdings, the stock comes to rest where injection and withdrawal balance, at M = N·GI/d, here 576.9 million; calibrated to that value, it holds there. The level is set by GI and d, not by the volume of transactions — the property the demurrage-tax runs will later turn on, since the base that tax draws upon is fixed here, by construction, and the bound is in place before a single unit of tax is collected.
The wealth distribution also compresses, with no revenue redistributed at all. The wealth Gini falls from 0.61 at calibration to 0.43, and the income Gini settles at 0.36. The mechanism is structural rather than fiscal: demurrage falls on balances in proportion to their size, so in absolute terms it draws most from the largest holdings, while the guaranteed income enters equally per head. The two together act as a continuous, mildly levelling force on the stock — enough to halt and partly reverse the concentration the underlying exchange dynamic generates by itself, the unconstrained form of which runs all the way to oligarchy in the baseline run, but not enough to equalise it. This is redistribution of wealth, but redistribution worked through the design of the money rather than through taxation.
That distinction is what the scenario is here to establish. The movement from 0.61 to 0.43 is the substrate's own contribution; whatever further compression appears once a fiscal layer is added — whether the revenue is raised by a demurrage tax, a tax on output, or income brackets — is the contribution of redistributing that revenue per capita, and it is measured against the 0.43 reached here. The paper's central distinction is already visible in this single run: a structural redistribution of the wealth stock is in place through the monetary design alone, prior to, and separable from, any redistribution of an income flow through the fiscal system.

The proposed configuration adds a fiscal layer to the substrate of the previous run, but not one of the familiar kind. Instead of taxing income or output, it levies a charge on the wealth stock itself — a demurrage tax, applied to holdings at 3.78 per cent a month alongside the base demurrage already in place. Its proceeds fund the residual obligation of 261 million, and fund it in full: the run raises 262 million and the budget balances, with the public debt flat across the horizon. Because those proceeds are spent back into the economy rather than withdrawn, the money stock is unaffected and holds at the same N·GI/d bound the substrate established on its own.  

The distribution this produces — a wealth Gini of 0.27 and an income Gini of 0.31 — is a marked compression on the 0.43 the substrate reached before any revenue was redistributed, and well below anything the conventional arrangements reach. Whether that further movement is owed to the demurrage tax in particular, or simply to returning the revenue to agents in equal shares, is left to the next run to settle. Here it is enough that the proposed system funds itself in full, holds its money stock at the bound, and compresses the distribution far past what taxing the income flow achieves.  

The monthly rate overstates what holders bear, and the gap is the first feature on which the case rests. The revenue the charge raises is returned to holders in equal per-capita shares, so what each holder actually carries is the charge net of an equal rebate — zero at the average balance of about 57,000, a net receipt below it, and a net charge above, approaching the gross rate only for holdings many times the average. The guaranteed income reaches everyone on the same flat basis and pulls in the same direction. A charge that scales with the balance, set against transfers that do not, is borne progressively — and it reaches that progressivity by construction, with no brackets and no thresholds, which is exactly what an income tax must assemble by hand and, as the later runs show, still cannot carry far enough.  

The second feature is the base. The charge falls on the stock and not the flow — the same 261 million an output tax raises by claiming roughly half of every year's income is drawn here from holdings, with the income flow left untaxed — and that stock is fixed, by the construction of the previous run, at N·GI/d. Its yield is therefore calculable from the bound and, as the runs under falling activity will show, does not move with the level of output. Neither feature shows up in the Gini, which is why the argument for the instrument is not made there; both become decisive once the economy is no longer held at full activity.

The case for the demurrage tax is therefore not, in the first instance, a distributional one: against a bounded stock it levels the distribution no further than a conventional output tax does. Under it the two distributions part company. The income distribution is held tight, its Gini settling near 0.24, because an equal transfer does to the income flow exactly what such a transfer does: it compresses it. This is the central result of the conventional runs. The equal redistribution governs the flow it is levied on and leaves the stock beneath it untouched: an income transfer, however large — and at basic-income scale it is large — redistributes income and does not redistribute wealth. What it buys is time, not a different outcome. The wealth distribution here is travelling to the same place the unconstrained baseline reached; the income floor only slows the journey, by keeping the poorest replenished, while the destination is unchanged. It is the monetary substrate, not the fiscal instrument, that alters that destination, as the bounded-stock runs show. And because this run is balanced and its money stable, the concentration cannot be laid to deficits or to monetary expansion — it is simply what redistributing a flow does, and fails to do, to a stock. The gap it leaves, a wealth Gini of 0.84 above an income Gini of 0.24, runs in the same direction as the gap between wealth and income inequality in the economies the model is drawn from.

For a time the tax keeps pace. The reason is that the tax is tethered to the wrong quantity. Its base is the income flow, which contracts with the economy, while the obligation it must fund does not; as output falls the base shrinks beneath a fixed target, and once the rate cap binds there is no scaling that closes the gap. A conventional income tax is, in this sense, pro-cyclically fragile: its capacity to fund the state erodes precisely as the economy weakens, which is when that capacity is most needed. The demurrage tax, set against the same decline, behaves in the opposite way, because the stock it draws on does not move with output — and that is the run the next section turns to, the contrast it draws being the central result of the exercise.

## 7 Discussion  

### 7.1 What the simulation establishes, and what it does not
The simulation results reported in Section 6 are mechanical and quantitative. Under unmitigated yard-sale dynamics the economy concentrates to a single holder within roughly a decade and ceases to function, reproducing the wealth-condensation result of Boghosian et al. (2015). Under conventional flow-coupled redistribution calibrated to Belgian fiscal capacity, the system stabilizes only provisionally: inequality drifts upward over centuries, and government operations are sustained in part by deficit spending, as they are in the present-day Belgian economy. Under AMS with a demtax calibrated to the residual expenditure, the system reaches a stable steady state with substantially lower inequality, a money stock bounded at the value the mechanism predicts analytically, and no reliance on deficit spending. These are findings about the dynamics of the system, established by the model and not contingent on any assumption about how agents respond psychologically to the conditions the mechanism creates.  

That boundary is worth stating plainly, because the discussion that follows crosses it. The yard-sale model contains no agent deliberation: agents do not decide, prefer, or respond: they transact according to fixed rules. The mechanical results therefore establish what the architecture does to the distribution of money under a given interaction process; they establish necessary conditions for the broader claims made on behalf of AMS, not those claims in full. Whether real agents, embedded in an AMS environment, would behave in ways that reinforce or undermine the mechanical results is a separate question that the present methodology cannot answer. The implications discussed in 7.2 through 7.4 follow mechanically from the results. The behavioural possibilities discussed in 7.5 do not; they are framed as hypotheses for methodologies better suited to testing them, supported at present by the existing literature on behaviour and context and by one small empirical study.  

### 7.2 Wealth comes to track income
The most consequential mechanical finding concerns the relationship between the wealth and income distributions. Section 1 noted that wealth inequality is consistently higher than income inequality and appears to operate more persistently, decoupling from income over time. In the simulation this decoupling is visible in the conventional scenarios: under flow-coupled redistribution the ratio of the wealth Gini to the income Gini sits near 1.96, close to the value of roughly 2.2 observed in Belgian data, and the wealth Gini continues to drift upward while the income Gini remains stable. Wealth concentrates independently of income, as the Matthew and r > g dynamics of Section 3 would predict.  

Under AMS with taxation the ratio collapses to 1.08: the wealth distribution becomes very nearly a scaled image of the income distribution rather than an amplification of it. The mechanism is the convergence property derived in Section 5. Each account tends toward telo(I, d) = I/d, an equilibrium balance proportional to net periodic income. Wealth therefore cannot detach from income and compound on its own, because any balance above the income-justified equilibrium is eroded by demurrage faster than it is replenished. The continuous downward pressure on accumulation that the mechanism applies is precisely a pressure against the decoupling of wealth from income. Where conventional redistribution offsets the symptoms of that decoupling after the fact, taxing accumulated wealth or high incomes through instruments that must be maintained against the underlying drift, AMS removes the conditions under which the decoupling occurs: wealth that is not continuously justified by income does not persist.  

This reframes what the intervention is doing. It is not, in the first instance, taking from the wealthy and giving to the poor, which is the logic of redistributive transfer. It is altering the relationship between two distributions so that one can no longer pull away from the other. The distributional compression observed in the simulation is a consequence of that altered relationship rather than of a transfer calibrated to produce it, which is why it is stable without ongoing adjustment: there is no drift for a maintenance effort to push against.  

### 7.3 Stock-coupled fiscal revenue
Under conventional arrangements the money stock grows. In the simulation it grows through sustained deficit spending modelled as money creation, which simplifies the real picture in two ways: most money creation in real economies occurs through commercial bank lending rather than through government deficits, and the deficits that are run are in substantial part financed by bond sales to non-bank investors, which move existing money rather than create it. The simulation's mechanism therefore overstates the money a given deficit brings into being. The growth it produces is nonetheless slower than the realized expansion of Belgian M3, which rose roughly threefold between 2001 and 2025, so the conventional scenario remains a conservative representation of the monetary environment despite the unrealistic channel through which its money stock grows. What distinguishes AMS is that its money stock is bounded by design, converging to the analytically predicted value |A_GI| · GI/d and remaining there — a convergence the simulation reproduces to the precision of the arithmetic.  

A further consideration concerns timing. The simulation collects demurrage and demtax and deposits the guaranteed income at synchronized period boundaries, so that the money stock sits exactly at its bound at each boundary. A real implementation would have a lag between collection and the spending that returns money to circulation, and the money stock would fluctuate around the aggregate telo rather than resting precisely on it. This is not a source of instability. The bound is a stable attractor and the demurrage mechanism is a proportional restoring force: a stock that runs temporarily below the telo is pushed back up by the next guaranteed-income deposit, and one that runs above it is pulled back down by the next demurrage charge. Deviations are corrected continuously rather than accumulating, and their amplitude is bounded by the size of a single period's flows — the difference between what is collected and what is spent. The bounded money stock is therefore a property of the mechanism rather than an artifact of synchronized accounting, and realistic timing lags would leave it materially unchanged.  

Because the money stock is bounded and its value is known in advance, the revenue a demtax will raise can be calculated before it is levied. A demtax rate d_t applied to a money stock of |A_GI| · GI/d yields a revenue of d_t · |A_GI| · GI/d, a quantity fixed by the policy parameters rather than emergent from economic activity. This is the property that makes the residual-expenditure problem tractable: the rate required to fund a given level of expenditure can be solved for directly, and the simulation confirms that the rate so derived covers the residual exactly. Conventional flow-coupled instruments do not have this property. The revenue from an income tax or a value-added tax depends on the level of incomes earned and transactions made, which are outcomes of the economy's behaviour rather than quantities the fiscal authority sets; the rate must be chosen in expectation of a revenue that is only known after the fact, and adjusted when the expectation proves wrong.
The same property makes demtax revenue independent of the level of economic activity. Because the revenue depends only on the bounded money stock and not on the volume of transactions, it neither rises in expansions nor falls in contractions. This is the inverse of the conventional case, where flow-coupled revenue contracts in a downturn — precisely when demands on the fiscal system rise — producing a procyclical squeeze the architecture must manage through borrowing or retrenchment. Demtax revenue is the same in a contraction as in an expansion. This stability carries a corollary that should be stated plainly: because the revenue does not grow with the economy either, an expenditure requirement that rises over time, through an expansion of public provision – an expansion of population would raise the money stock and therefore the collected demtax – would eventually outrun a fixed demtax rate, and the rate, and/or or the guaranteed income that together sets the size of the money-stock base — would have to be adjusted upward to match. Demtax revenue is therefore insulated from economic fluctuation but not self-adjusting to secular changes in expenditure need; it trades the automatic growth of flow-coupled revenue for stability against the cycle.  

This trade-off is of particular relevance to the steady-state and degrowth perspectives within ecological economics. Conventional fiscal systems are growth-dependent in a structural sense: flow-coupled revenue requires economic activity to sustain it, and the sustainability of public debt depends on growth outpacing the cost of servicing it. In an economy that is not growing, whether by intention or by circumstance, these dependencies become stresses, as revenue stagnates and debt service consumes a rising share of it. A fiscal instrument whose revenue is decoupled from the volume of activity does not carry these dependencies: demtax provides stable revenue in a steady-state economy without requiring growth to underwrite it, and the incentive that allows bonds with a negative interest removes the compounding debt-service dynamic entirely. AMS thus offers a fiscal architecture that does not presuppose growth, a property of direct interest to a research programme concerned with whether a non-growing economy can be made viable. The dynamic behaviour of the system under sustained growth or contraction lies beyond the scope of the steady-state simulations reported here and is left to future work; the independence of demtax revenue from the level of activity, however, is a structural feature of the mechanism, derivable independently of any growth scenario.  

### 7.4 AMS as a structural counter-loop
Section 4 distinguished interventions that offset the outputs of the inequality-generating loops from interventions that alter the loops themselves, and placed conventional redistribution in the first category: it transfers wealth from the top of the distribution to the bottom while the dynamics that produced the concentration continue to operate, which is why it must be maintained against a drift it does not arrest. The results of Section 6 indicate that AMS belongs to the second category. It does not offset the concentration dynamic; it introduces a dynamic of opposite sign. The demurrage applies a downward pressure on monetary holdings proportional to their size, which is the structural inverse of the wealth-attained advantage that drives concentration in the yard-sale model: where the latter is self-reinforcing, channelling wealth toward those who already hold it, the former is self-attenuating, eroding holdings in proportion to how far they exceed the level current income sustains. The stabilization observed in the simulation is the visible result of this opposing dynamic, and its persistence without maintenance — the absence of the upward drift the conventional scenario exhibits — is the signature of a counter-loop rather than an offset.  

Several of the loops identified in Section 3 are countered by the structure of the mechanism. The yard-sale concentration dynamic is countered at its source, as the simulation directly demonstrates: the same interaction process that drives the baseline to oligarchy produces a stable distribution once demurrage is present. The Matthew effect and the r > g dynamic, which depend on held wealth compounding independently of the income that produced it, are countered by the telo coupling examined in 7.2: under demurrage, idle monetary wealth depreciates rather than appreciates, so wealth that is not continuously replenished by income does not persist, and the passive accumulation those loops require cannot occur. Wealth can still be built through income, but it can no longer build itself. The credit-money loop is addressed by the architecture rather than by the simulation, which contains no banking sector: where credit-money creation channels new money toward those able to post collateral — that is, toward existing wealth — AMS creates money through the guaranteed income and distributes it equally, while bounding the resulting stock. This inverts both the distributional bias of money creation, which in the conventional architecture is a vector of concentration and under AMS a vector of dispersal, and the instability that Section 3, following Kuypers et al. (2021), attributes to credit-driven monetary expansion.  

The remaining loops of Section 3 — the bargaining asymmetry between rich and poor parties, the differential vulnerability to shocks, the conversion of wealth into political influence, and the ideological reinforcement of inequality through meritocratic belief — are not represented in the simulation, and AMS does not act on them directly. What the simulation establishes is that AMS compresses the wealth distribution and ties it to income, and each of these loops is driven, on the analysis of Section 3, by the concentration of wealth: the bargaining asymmetry widens with the wealth gap between parties, vulnerability to shocks tracks the absence of a buffer, political influence follows the concentration of resources, and meritocratic belief intensifies with material inequality. To the extent that these relationships hold, compressing the distribution weakens the driver common to all four. But the loops themselves operate through behavioural and institutional channels the model does not contain, so the claim that AMS weakens them is a claim about a necessary condition — the reduction of the concentration that drives them — rather than a demonstration of the downstream effect. Whether the compression observed in the simulation would in fact translate into reduced bargaining asymmetry, lower shock vulnerability, diminished political capture, and weaker meritocratic legitimation is a question for methods able to represent those channels, of the kind discussed in 7.5.  

One feature of the counter-loop bears directly on the loop that conventional redistribution is least able to address. The political-influence loop attacks redistributive policy at its point of maintenance: because parameter level redistribution must be continuously renewed through political decision, the concentration of influence among those with the most to gain from weakening it makes that maintenance progressively harder to sustain. A counter-loop operating at the level of monetary policy — parameterized by the guaranteed income and the demurrage rate, set in the manner of monetary rather than fiscal policy — does not present the same recurring point of attack, because the redistributive effect is produced by the standing operation of the mechanism rather than by a transfer that must be re-authorized each period. This does not place AMS beyond political reach: the parameters can be altered, and the architecture itself is a political choice. But it changes what sustained influence would have to accomplish — not the gradual erosion of a maintenance effort, but the structural revision of the monetary system. AMS is, in this sense, not a larger offset to the dynamics of Section 3 but an intervention of a different kind, one that installs a generative dynamic of opposite sign and thereby alters the conditions those dynamics require in order to operate.  

### 7.5 The transfer and the counterforce: AMS and the basic-income question
A natural reading of AMS attributes its distributional effect to the guaranteed income alone — to the unconditional transfer, with the demurrage doing no more than holding the money stock bounded. The scenario set includes two conventional runs without AMS, and the one that tests this reading is the run in which a flat income tax covers government expenditure in full, with no deficit. The equal redistribution of that tax revenue is, by construction, a tax-funded basic income: the entire tax take is returned as an identical per-capita payment, so the transfer each account receives equals the total revenue divided across the population — 26,130 per year, or 2,178 per month — and because the budget is balanced and the stock flat, this is precisely the tax recycled per head. Whether this equal return is read as a basic income or as government revenue returned to households makes no difference to what follows, since both are flow transfers; the basic-income reading is simply the most favourable case for redistribution, a uniform per-capita payment being the most compressing form the return could take.  

What it accomplishes is the stabilization of income, not of wealth. The income distribution is compressed and held — its Gini settles near 0.24 — and no account is left destitute in flow terms; even the lowest wealth decile retains an income on the order of 35,000 against a population mean near 51,000. The wealth distribution does the opposite. After an initial dip to about 0.50 over the first decade, the wealth Gini climbs to 0.83 within a century and approaches near-total concentration over the longer run, the top decile's share rising past four-fifths and toward the whole, the bottom half left with under a tenth. And it does so while the money stock remains flat: the concentration cannot be charged to monetary expansion, because there is none. The equal transfer governs the variable it pays out and leaves the variable that concentrates — the stock of accumulated balances — to follow the unmitigated exchange dynamics of Section 3.  

The reason is structural, and it is the distinction between a flow and a stock. Wealth concentration in this setting is a property of the stock: it arises from the repeated, multiplicatively biased exchange of accumulated balances, in which the better-capitalized side has the greater staying power. A transfer that adds an equal amount to every account's inflow raises the floor and slows the drift, but supplies no force that scales with the stock it is meant to counter; the balances it replenishes at the bottom are drawn back through the same exchange, while those at the top compound faster than a levy on the income flow can recover. Demurrage is exactly the term the basic income lacks — a charge proportional to the balance itself, whose restoring pressure rises with the wealth it acts on. This is the same role a wealth-proportional redistribution term plays in the kinetic-exchange literature, where it sets the condition under which condensation is averted rather than merely postponed (Boghosian et al., 2017). It is the demurrage, and not the transfer, that converts the system from one that delays concentration into one that reaches a stable distribution: with it in place the wealth distribution stabilizes whether the fiscal layer above it is levied on wealth or on income, and without it the same flat levy and equal transfer concentrate as the bare dynamics dictate — the flat money stock confirming that nothing but those dynamics is at work.  

This locates the contribution of AMS precisely, and answers the question the architecture invites: it is not a basic income under another name — and not because it lacks one. AMS contains a guaranteed income, itself an equal per-capita flow transfer of exactly the kind tested here, distinguished only by being funded through money creation rather than tax. The telo, GI/d, divides the labour between the two: the guaranteed income, in the numerator, sets the level at which an account settles, while the demurrage, in the denominator, supplies the convergence to it. A flow transfer by either funding route fixes the floor and the level; it is the demurrage that makes the distribution stable. The two are complements, and an unconditional transfer alone — whether the tax-funded one tested here or the money-funded guaranteed income considered in isolation — is necessary but not sufficient for distributional stability: sufficient to secure incomes, insufficient to govern the wealth stock. Two qualifications bound the claim. The first is the funding source: a basic income financed by a levy on wealth rather than on income would itself carry a stock-scaling counterforce, and in the limit a continuous wealth levy redistributed as an equal payment simply is a demurrage paired with a guaranteed income — so the operative distinction is between flow-funded transfers with no stock instrument and arrangements that include one, not between basic income and AMS as categories. The second is time: the full-coverage transfer delays concentration by a large factor — centuries rather than the bare model's years — so over policy-relevant horizons its effect on inequality is real, and the insufficiency established here is asymptotic, in a model that abstracts from labour income, productive investment, and inheritance. Within those bounds the point stands: a flat-tax-funded basic income stabilizes the distribution of income but not of wealth, because it acts on the flow and not on the stock.

### 7.6 Behavioural responses the mechanism makes possible
The implications discussed so far follow from the mechanical results. The remainder of this section turns to a question the simulation cannot address: how agents embedded in an AMS environment might behave. The agents of the yard-sale model do not deliberate; they transact by fixed rules, and the distributional outcomes reported above are properties of those rules rather than of any choice. Real agents would meet the conditions the mechanism creates as a situation to respond to, and their responses would shape outcomes the model holds fixed. What follows is a set of hypotheses, grounded in the relationship between behaviour and context developed in Section 4.1 and in the single empirical study available, and offered to motivate the methods that could test them.  

The most direct question concerns holdings above the telo. An account above its convergence point loses more to demurrage than it receives in income, and the holder faces a choice about money that would otherwise erode: to spend it, to deploy it in a way that generates income, to give it away, or to hold it and absorb the loss. The mechanism does not dictate which of these occurs; it changes the relative cost of the last. Under conventional conditions idle monetary wealth is nominally preserved or grows, so holding is a viable option against risk bearing mobilization; under demurrage holding carries a continuous cost, thus making mobilization the more viable option. This is a change in the choice architecture rather than in the agents who face it. It does not require holders to become less acquisitive or more generous: an acquisitive holder who wishes to grow wealth must, under AMS, deploy money in ways that generate income, because monetary wealth held idle no longer grows. The same disposition that produces hoarding under one architecture produces circulation under the other. This is the sense, developed in Section 4.1, in which the intervention works with the distribution of human dispositions as it is rather than requiring it to change. The mechanism does not, however, distinguish among the forms mobilization takes; whether money is moved into productive investment, consumption, or extractive activity is not something the demurrage determines, and the social value of the circulation it encourages is a separate question.  

The incentive structure at the other end of the distribution is also altered. Because the guaranteed income is unconditional, income earned through work or enterprise is not clawed back as it rises, and because equilibrium wealth tracks net income through the telo, earning more raises the level at which an account settles. An agent at the floor who seeks a higher standard of living therefore faces a positive return to earning, in contrast to the high effective marginal rates that means-tested benefits impose at the bottom of conventional systems, where additional earnings are offset by withdrawn support. Whether agents respond to this incentive, and how the response is distributed across the population, is an empirical matter; what the mechanism supplies is an incentive structure without a poverty trap, not a prediction about uptake. A related and more open question concerns the top of the distribution: whether agents possess a comfort threshold beyond which the incentive to accumulate weakens, and if so, how the resulting behaviour interacts with the aggregate dynamics. The bounded money stock caps aggregate accumulation wherever individual thresholds lie, but the shape of the distribution within that bound would depend on dispositions the model does not represent.  

The one body of empirical evidence bearing on these questions is a small gaming study (Haantjes, 2024) in which participants managed resources under conditions modelling an AMS-like environment. Participants shifted within a single session from transactional toward cooperative framings of their decisions, directed resources toward common rather than purely individual ends, and reported higher wellbeing and lower stress than under the conventional comparison condition. These results are consistent with the behavioural responses hypothesized above and with the broader claim that the conditions agents face shape the dispositions they express. They are, however, the results of one study with a small sample in an artificial setting, and the degree to which behaviour in a game transfers to behaviour in an economy is precisely what such a study cannot settle. The evidence is suggestive of a direction, not confirmatory of a magnitude.  

The behavioural layer is therefore the part of the case for AMS that this paper leaves open. The mechanical results establish what the architecture does on its own terms; whether agents embedded in it would behave in the ways the choice architecture makes available, and whether those behaviours would reinforce or erode the mechanical results, requires methods able to represent deliberating agents — behavioural experiments, larger and more naturalistic gaming studies, and agent-based models with richer psychology than the present one. The contribution of this paper is to establish the mechanical results and to mark clearly where their evidence ends and the behavioural evidence would have to begin.

### 7.7 A general principle
The argument of this paper has a structure that is not specific to inequality. A persistent social outcome is observed; it is found to be generated by reinforcing dynamics embedded in the rules of a system; interventions that adjust parameters within those rules, or that target the behaviour the dynamics produce, are found to offset the outcome without arresting it, and so to require continuous maintenance; and an intervention on the rules themselves is shown to alter the dynamics at their source. Inequality is a particularly clear case in which to demonstrate this structure, for two reasons. The dynamics that generate it are formally tractable, which allows the mechanical claims to be stated and tested precisely rather than asserted. And the inadequacy of parameter intervention is unusually well documented in this domain, in Piketty's observation that the compression of inequality in the mid-twentieth century required, and decayed without, sustained political effort against an underlying drift. The result reported here is, in this light, a worked instance of a more general proposition: that for outcomes generated by structural dynamics, structural intervention has higher and more durable leverage than the parameter and behavioural interventions that operate within the existing structure.  

If the proposition holds, its interest extends beyond inequality to other outcomes that have resisted behavioural and parameter intervention, two of which are of particular relevance to the concerns of ecological economics. The first is pro-social behaviour: the persistent gap between the cooperation that collective problems require and the cooperation that individuals supply has been addressed largely through appeals to disposition — education, moral suasion, the cultivation of norms — which, on the analysis of Section 4.1, target the behaviour rather than the context that produces it. The choice architecture examined in 7.5, in which the cost of holding rather than circulating money shifts the default away from accumulation without requiring any change in disposition, illustrates the structural rather than the dispositional approach; whether monetary or related structures can be designed to shift defaults toward cooperative rather than extractive activity, in the same way, is a question the present framework could be extended to examine. The second is regenerative, as against extractive, economic activity. The central concern of ecological economics is an outcome — the depletion of natural systems — generated by dynamics embedded in the structure of the economy, and addressed predominantly through parameter interventions, taxes and subsidies and regulatory thresholds, that must be maintained against the drift they counteract. Whether a monetary architecture can be designed so that regenerative activity is structurally favoured, in the way that AMS structurally disfavours accumulation, is the natural extension of the present argument to the field in which it is offered. These are directions for research, not results: the present paper establishes the principle in one domain and does not demonstrate its application to others.  

One qualification must accompany this generalization. Higher leverage is not lower difficulty. The same property that makes structural intervention durable — that it changes the rules within which dynamics operate rather than the magnitudes those dynamics produce — makes it harder to enact, since it requires altering arrangements that are entrenched precisely because they are structural. The claim advanced here concerns where intervention has the most durable effect, not where it is easiest to achieve; and the political-influence loop examined in 7.4 suggests that the durability and the difficulty share a common source, in that the arrangements with the highest leverage are also those their beneficiaries have the most reason and the most capacity to defend. The contribution of this paper is to establish, in a single tractable domain, that the leverage is real — that a structural monetary intervention produces, mechanically, the stable and bounded outcomes that parameter intervention produces only provisionally and at the cost of perpetual maintenance — and to set out the terms on which the same approach might be brought to bear on the wider class of problems that share its structure.

## 8. Limitations
The simplifications adopted in the simulation fall into two kinds, and it is useful to separate them.  

The first are simplifications whose effect is conservative with respect to the argument: they make the conventional scenario appear more favourable, or the AMS scenario less favourable, than a fuller model would, so that the advantage attributed to AMS is if anything understated.  

The second are genuine constraints on what the study can claim, whose direction is neutral or uncertain. The interpretive limits of the study — the boundary between the mechanical results the simulation establishes and the behavioural responses it cannot — are set out in Section 7 and are not restated here. Several features of the monetary and fiscal modelling understate the contrast the paper draws. The money stock in the conventional scenario grows only through deficit spending treated as money creation, which both omits the dominant real channel of money creation, commercial bank lending, and overstates the money a deficit creates, since real deficits are in substantial part financed by money-neutral bond sales to non-bank investors. The resulting growth is nonetheless slower than the realized expansion of Belgian M3 over the past quarter-century, so the contrast between a growing and a bounded money stock is, if anything, muted relative to what a fuller monetary model would show (Section 7.3). Interest on public debt is not modelled in the conventional scenario, although in real economies debt-service payments flow disproportionately to the holders of government debt, who sit toward the top of the wealth distribution; including this channel would add a concentration mechanism to the conventional scenario that the simulation lacks. And the AMS scenarios are funded entirely by demtax, with no recourse to the negative-cost bond financing the architecture permits (Section 5.2); this is the more demanding assumption, since it requires the demtax to cover the full residual expenditure unaided and credits AMS with none of the fiscal flexibility that borrowing at or below zero interest would provide.  

A further conservative feature concerns the baseline. The conventional scenario settles at a more equal distribution than is observed in Belgium, so the advantage AMS shows over it is smaller than the advantage it would show over the more concentrated distribution real arrangements produce. The reason for this gap is not that the flat tax is more redistributive than the system it stands in for — a flat tax is, if anything, less progressive than a tiered schedule. A more likely explanation, that is not verified directly, lies in how the tax is applied: the simulation levies a single rate on the whole of GDP with no exemptions, deductions, or avoidance, whereas real tax systems leak through all three, particularly at the top. The simulated tax, though nominally flatter, may therefore capture a larger share of economic activity for redistribution than the real system does, producing a more equal outcome in spite of its lower progressivity. What matters for the comparison is the direction of the effect rather than its precise source: the simulated baseline is more equal than reality, so the advantage attributed to AMS is, on this count too, conservative.  

The remaining limitations constrain what the study can claim without biasing the comparison in a determinate direction.  

The population is homogeneous, with no age structure or other demographic heterogeneity; this isolates the system dynamics from demographic complexity at the cost of not representing the distributional effects of age, household structure, or life-cycle position.  

Agents interact on a fully connected network, in which any agent may transact with any other. The networked variant of the extended yard-sale model shows that this assumption produces global rather than local wealth condensation and represents an upper bound on concentration relative to realistic interaction structures (Vazquez Von Bibow and Perotti, 2025); the quantitative magnitudes reported here may therefore differ on a sparse network, though the qualitative phase-transition structure underlying the results is preserved across both.  

The wealth-attained advantage is a single reduced-form parameter standing in for several distinct concentration mechanisms — returns to capital, inheritance, differential access to credit, and the avoidance of taxation among them — which are not modelled individually but absorbed into its calibration; the model captures the net concentrating force these mechanisms exert without distinguishing their separate contributions.  

Several narrower choices should also be noted.  

The wealth-attained advantage is expressed in a per-transaction formulation whose calibrated value is specific to the simulation's GDP-calibrated transaction architecture and is not directly comparable to the values reported in the continuous-time econophysics literature, which use a different scaling convention; the value should be read as an internal calibration rather than an estimate of a quantity reported elsewhere.  

GDP serves as the reference against which simulated time is measured. Property taxation, a stock-coupled instrument accounting for a small share of Belgian revenue, is omitted, so the conventional fiscal mix is represented as purely flow-coupled. The economy is modelled as closed; while every national economy is open, the global economy is closed, and the flat tax is calibrated to reproduce Belgian fiscal magnitudes within that closed frame. The use of money as the operative variable, defended in Section 4.3, means that non-monetary and non-liquid wealth is represented only at the point at which it is mobilized into monetary form, so accumulations that are never mobilized lie outside the dynamics the model tracks. Finally, a tax-collection edge case arises at extreme concentration, where a large majority of accounts hold negative balances and the fallback in the collection routine behaves incorrectly; this does not affect any of the steady-state results reported in the paper, which lie far from that regime, but it is disclosed because it bears on the interpretation of runs carried to extreme concentration.Taken together, the simplifications of the first kind suggest that the advantage the paper attributes to AMS is conservative, while those of the second mark the boundaries within which that advantage has been established: a homogeneous, fully connected, closed economy whose concentration dynamics are summarized in a single calibrated parameter. The results hold within those boundaries; extending them — to heterogeneous populations, realistic interaction networks, open economies, and explicitly modelled concentration mechanisms — is the work that would show how far they generalize.

A further simplification concerns how collected revenue is returned. The model redistributes both the guaranteed income and the demtax proceeds as an equal per-capita amount, and this even return is part of what produces the demtax's progressive incidence: since every account receives an identical share, the net charge is negative below the mean balance and positive above it, crossing zero at the mean. Real fiscal systems do not return revenue as an equal cash rebate; they fund public provision whose incidence is uneven and, in some components, deliberately concentrated on lower-income households. The direction of the resulting bias is not determinate — an equal per-capita return is the most compressing form redistribution can take, so real spending that is less even would make the net incidence less progressive than the model shows, while spending targeted down the distribution would make it more so. The activity-independence of the demtax's gross yield, on which the paper's fiscal argument rests, does not depend on the return rule; but the distributional outcomes reported for the demtax runs are conditioned on it, and should be read as the cleanest case rather than a forecast of a particular programme's net effect.

## 9. Further research
The results established here are mechanical, and the questions they leave open fall into several distinct programmes. Most have been flagged at the points in the discussion where they arise; this section gathers them and develops the ones the body does not.  

The largest is behavioural. The simulation fixes agent behaviour by rule, so what it establishes is what the architecture does to the distribution of money under a given interaction process, not how agents who could deliberate would act within it (Section 7.6). Characterizing the behavioural layer would require behavioural experiments, gaming studies larger and more naturalistic than the single one available (Haantjes, 2024), and agent-based models with richer psychology than the present one. Two narrower threads sit inside this programme. The first concerns the use of money itself as an experimental instrument: experiments routinely use monetary cues as rewards and punishments to elicit responses in moral-choice settings (Kang and Camerer, 2013), but if the behavioural force money carries is itself a function of the monetary architecture — as the argument of Section 4.1 implies — then results obtained under conventional money may not transfer to an architecture in which holding money carries a continuous cost, and whether non-monetary rewards reproduce the same effects becomes a question worth revisiting. The second concerns the evidence from complementary-currency schemes, among whose users elevated pro-social behaviour has been observed: whether this reflects the design of those currencies or the dispositions of the people who adopt them is unresolved, and it is the same confound — design versus selection — that bounds what the gaming study here can show, since its participants were not randomly assigned to the monetary environment they experienced. Disentangling the two requires designs in which the architecture is assigned rather than chosen.  

A second programme concerns the step from compression to consequence. The simulation establishes that AMS compresses the wealth distribution and binds it to income; the harms that motivate the exercise — the bargaining asymmetry, shock vulnerability, political capture, and meritocratic legitimation of Section 7.4 — are driven by concentration but operate through channels the model does not contain, so whether the compression observed here would in fact attenuate them is a question for methods able to represent those channels. A more basic question concerns the target itself. The relationship between the measured level of inequality, the level a population perceives, and the outcomes associated with either is neither linear nor well characterized, and the perceived level may carry more weight than the measured one (Norton and Ariely, 2011; Gimpelson and Treisman, 2018; Du and King, 2022). Establishing whether thresholds exist — levels below which the associated harms measurably abate — would allow the architecture's parameters, the guaranteed income and the demurrage rate, to be set against targets defined by their effect on those outcomes rather than against the goal of reducing measured inequality as such. Such thresholds are likely contingent on cultural and ideological context, and the neuroscience of fairness perception suggests they are not reducible to the measured distribution alone (Cappelen et al., 2014); the work is therefore interdisciplinary, drawing on psychology, sociology, and neuroscience as much as on economics.  

A third programme concerns how far the mechanical results extend beyond the simplifications under which they were obtained. The boundaries are set out in Section 8: a homogeneous, fully connected, closed economy whose concentration dynamics are summarized in a single reduced-form parameter, and which abstracts from labour income, productive investment, and inheritance. Relaxing these — introducing demographic and household heterogeneity, sparse and realistic interaction networks, on which concentration is bounded below the fully connected case (Vazquez Von Bibow and Perotti, 2025), open rather than closed economies, and concentration mechanisms modelled explicitly rather than absorbed into one parameter — would show how much of the compression result survives richer structure. The dynamics of the mechanism under sustained growth, and under a secular rise in expenditure need that a fixed demtax rate would not automatically track (Section 7.3), likewise remain open: the contraction case is addressed by the declining-output runs reported above, but the symmetric growth case, and the transitional dynamics of a money stock moving from one architecture toward another, are not.  

A fourth programme is distinct in kind. The paper establishes that the leverage is real but says nothing about the route to realizing it. A structural intervention alters arrangements that are entrenched precisely because they are structural, and, as Sections 7.4 and 7.7 observe, the arrangements with the highest leverage are those their beneficiaries have the most reason and the most capacity to defend; the durability that recommends the intervention and the difficulty that obstructs it share a source. How such an architecture might be introduced, given existing power dynamics and the inequalities the current system has already produced, is a research programme in its own right, spanning political economy, institutional and legal design, and the study of partial or parallel implementations — complementary-currency pilots, regional or sectoral schemes — as possible transitional steps. It is the question on which the practical force of the result depends, and the one the present paper is least equipped to answer.

## 10. Conclusion
This paper asked whether the persistent upward drift of inequality — which conventional redistribution offsets but does not arrest — might be addressed at the level that generates it, in the rules of the monetary system rather than in the magnitudes of transfers layered on top. It examined a candidate architecture, AMS, that pairs an unconditional guaranteed income with a demurrage on monetary balances and funds the state through a demurrage tax, and tested it in an agent-based model calibrated to Belgian monetary and fiscal magnitudes, against a baseline of unmitigated wealth-exchange dynamics and a conventional flow-coupled redistributive scenario.  

Within the model the results are clear. Unmitigated dynamics drive the economy to complete concentration within years. Conventional flow-coupled redistribution, calibrated to Belgian fiscal capacity, holds inequality down only provisionally, leaving it to drift upward toward observed Belgian levels over the long run while the state sustains itself through continuous deficit spending. The AMS architecture reaches a stable steady state instead: wealth inequality settles at a Gini of 0.26 and remains there, the money stock converges to the bound the mechanism predicts analytically and stays within it, the demtax covers the residual government expenditure exactly — at a rate calculable in advance and independent of the level of economic activity — and no public debt accumulates. The wealth distribution comes to track the income distribution rather than detaching from it, the ratio between the two compressing to close to one. These outcomes are produced without continuous maintenance, because the mechanism alters the dynamics that generate inequality rather than offsetting their results; and as the simplifications of the model are, on the analysis of Section 8, largely conservative, the advantage they show is more plausibly understated than overstated.  

These are mechanical findings, established by the model and not contingent on how agents respond to the conditions the architecture creates. Whether real agents would behave in ways that reinforce or erode them is a question this study delineates but does not answer, and one better suited to the experimental and behavioural methods discussed above. What the study does establish is that the leverage is real: in a single tractable domain, a structural intervention on the monetary system produces, on its own terms, the stable and bounded outcomes that parameter intervention reaches only provisionally and at the cost of perpetual effort. Whether the same approach can be brought to bear on the wider class of problems that share this structure — among them the regenerative transformation that is the central concern of ecological economics — is the question the result opens.
